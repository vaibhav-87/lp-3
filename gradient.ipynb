```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = 2
learning_rate = 0.01
precision = 0.000001
max_iterations = 10000
```

```python
def slope(x):
    return 2 * (x + 3)

def y_value(x):
    return (x + 3) ** 2

x_values, y_values = [x], [y_value(x)]
```

```python
for iteration in range(max_iterations):
    new_x = x - learning_rate * slope(x)
    x_values.append(new_x)
    y_values.append(y_value(new_x))
    print(f"Iteration {iteration}: X = {new_x:.6f}")
    
    if abs(new_x - x) < precision:
        print(f"Converged at iteration {iteration + 1}")
        break
        
    x = new_x

print(f"Local minimum occurs at: {x:.6f}")
```

```python
sns.scatterplot(x=x_values, y=y_values)
```
